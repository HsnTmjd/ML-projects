```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset = pd.read_csv('/content/boston.csv')
```

```python
print(house_price_dataset)
```

```python
house_price_dataset.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
house_price_dataset.head()
```

```python
house_price_dataset.describe()
```

understanding the correlation between various features in the dataset

1. positive correlation
2. Nagetive correlation

```python
# finding the co-relation
correlation = house_price_dataset.corr()
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Reds')
```

```python
 # spliting the data
 X = house_price_dataset.drop(['PRICE'],axis=1)
 Y = house_price_dataset['PRICE']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

Evaluation

Prediction on the training data

```python
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

Now we are getting the error value by comparing the prediction with Y_train values

456554736_1661156904678177_7453052347265361532_n.png

```python
# R squared error
score_1 = metrics.r2_score(Y_train,training_data_prediction)
print('R squared error : ',score_1)
# mean absolute errors
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)
print('Mean Absolute Error : ',score_2)
```

Visualizing the actual price and predicted price

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

```python
accuracy_score = model.score(X_test,Y_test)
print(accuracy_score)
```

prediction on test data

```python
test_data_prediction = model.predict(X_test)
```

```python
# r squared error
score_1 = metrics.r2_score(Y_test,test_data_prediction)
print('R squared error : ',score_1)
# mean absolute errors
score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction)
print('Mean Absolute Error : ',score_2)
```

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```
